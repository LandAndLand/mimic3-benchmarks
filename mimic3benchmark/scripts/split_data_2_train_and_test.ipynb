```python
import os
import shutil
import argparse
import random

from tqdm import tqdm
```

```python
subjects_root_path = 'data/root'
model_data_dir = 'data'
all_patients_num = 33798
test_patients_num = 5070  # 15%
```

```python
subjects = []
for subject_dir in tqdm(os.listdir(subjects_root_path), desc='Iterating over subjects'):
    #print(subject_dir)
    subjects.append(int(subject_dir))
print(subjects)
```

```python
test_subjects =  random.sample(subjects, test_patients_num)
print(len(test_subjects))
train_subjects = [i for i in subjects if i not in test_subjects]
print(len(train_subjects))

assert len(set(test_subjects) & set(train_subjects)) == 0
```

```python

def move_to_partition(patients, partition):
    if not os.path.exists(os.path.join(model_data_dir, partition)):
        os.mkdir(os.path.join(model_data_dir, partition))
    for patient in patients:
        patient = str(patient)
        src = os.path.join(subjects_root_path, patient)
        dest = os.path.join(model_data_dir, partition, patient)
        #shutil.copytree(src, dest)
        shutil.move(src, dest)
```

```python
move_to_partition(test_subjects, 'test')
test_subjects(train_subjects, 'train')
```
